## 1. Concept of Split–Apply–Combine
- **Split** the data into groups based on some criteria.
- **Apply** a function (e.g., sum, mean) to each group.
- **Combine** the results into a new DataFrame or Series.

This is the core idea behind `groupby`.

```python
import pandas as pd

# New dataset: Employees, their Department, and Salary
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
df.groupby('Department')['Salary'].mean()  # har department ki average salary kya hai?
```

```python
count_employees = df.groupby('Department')['Employee'].count()   # Har department me kitne log hai?
print(count_employees)
```

```python
multi_stats = df.groupby('Department')['Salary'].agg(['mean', 'min', 'max'])
print(multi_stats)
```

## 2. Using `groupby()` with Aggregation Functions
- Use `groupby('col').mean()` or other functions like sum, count, min, max.
- You can also use `agg()` to pass multiple functions at once.

## 3. Multiple Aggregations on Different Columns
- With `agg()`, you can specify different functions for each column.
- Example: `df.groupby('Dept').agg({'Salary':'mean', 'Age':'max'})`.

```python
import pandas as pd

# Same dataset
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000],
    "Age": [28, 32, 30, 27, 35, 29, 31, 26]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Department wise average Age.

avg_age = df.groupby('Department')['Age'].mean()
print(avg_age)
```

```python
df.groupby('Department').agg(
    {'Salary' : ['mean','min','max'],
     'Age' : ['min','max']
    }
)
```

## 4. Pivot Tables & Crosstab
- `pivot_table()` summarizes data like Excel pivot tables.
- `crosstab()` shows frequency counts of combinations of factors.
- Both are powerful for summarizing and comparing groups.

```python
# df.pivot_table(index='konsa column main hai', values='Kya chahiye', aggfunc='mean','min','max','count')

df.pivot_table(index='Department', values='Salary', aggfunc='mean')  # for 1 cloumn

df.pivot_table(index='Department', values=['Salary','Age'], aggfunc='mean')  # for multiple column

# Age Group -> 21-25 , 26-30 , 31-35
df['Age_Group'] = pd.cut(df['Age'], bins=[25,30,35,40], labels=['25-30','31-35','36-40'])

pd.crosstab(df['Department'], df['Age_Group'])
```

```python

# 1. Create a DataFrame with columns: Department, Employee, Salary, Age, City
data = {
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000],
    "Age": [28, 32, 30, 27, 35, 29, 31, 26],
    "City": ["New York", "San Francisco", "San Francisco", "Los Angeles", "Los Angeles", "New York", "San Francisco", "Los Angeles"]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
# print(df)

# 2. Group by Department to get average Salary and max Age
# Write your code here
avg_salary = df.groupby('Department')['Salary'].mean()
# print(avg_salary)

max_age = df.groupby('Department')['Age'].max()
# print(max_age)

# 3. Apply multiple aggregations using agg()
# Example: mean Salary, max Age per Department
# Write your code here
df.groupby('Department').agg({'Salary':'mean', 'Age':'max'})

# 4. Create a pivot_table to see mean Salary by Department and City
# Write your code here
df.pivot_table(index='Department', columns='City', values='Salary', aggfunc='mean')

# 5. Build a crosstab for Department vs City
# Write your code here
pd.crosstab(df['Department'], df['City'])
```
