##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np

arr = np.array([10,20,30,40,50,60])

# Pick elements at position 0,2,4
print(arr[[0,2,4]])
```

You can do this fancy indexing in 2D Matrix aswell

```python
matrix = np.arange(1,13).reshape(3,4)
# matrix[[row1,col2],[row0,col2]]
# matrix[[0,1],[1,2]]
print(matrix)
matrix[[1,2],[0,2]]
```

##Conditional Selection
Find values based on conditions or replace them.

```python

arr = np.array([5,7,11,18,30])

# we will get the indices only which are greater than 10
np.where(arr > 10)

# Array me jaha bhi 10 se chhoti value aa rahi hai waha 0 se replace kar do

print(np.where(arr > 10, arr , 0))
```

```python
arr2 = np.array([23,25,30,50,100,75,80])

# Replce the value which are less than 55

print(np.where(arr2 > 55 , arr2 , 0))
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
# Aam Zindagi
array = np.array([2,10,56,1,54])
print(np.sort(array))

# Mentos Zindagi
index = np.argsort(array)
print(index)
```

```python
# Matrix Sort
matrix = np.array([[4,12,3],[14,10,15]])

#1. Row wise sort (axis = 1)
print(np.sort(matrix, axis = 1))

#2. Column wise sort(axis = 0)
print(np.sort(matrix, axis = 0))

# Sort Complete Matrix
print(np.sort(matrix))
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.concatenate([a,b])
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[4,5]])
B = np.array([[7,8],[10,11]])

# Vertical Stacking
print(np.vstack([A,B]))

# Horizontal Stack
print(np.hstack([A,B]))
```
