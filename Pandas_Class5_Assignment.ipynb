## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
# Write your code here
import pandas as pd
df = pd.DataFrame({'Age': [25, 18, 30, 22]})
df.sort_values(by='Age')
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
# Write your code here

df = pd.DataFrame({'Marks': [60,70,80,90]})
print('Mean =', df['Marks'].mean())
print('Median =', df['Marks'].median())
print('Max =', df['Marks'].max())
print('Min =', df['Marks'].min())
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
# Write your code here
df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D'],
    'Age': [20, 25, 30, 35],
    'Score': [80, 70, 90, 60],
    'Height': [170, 165, 180, 175]
})
print(df.sort_values(by='Score', ascending=False))
print(df['Age'].mean())
print(df['Height'].median())
print(df['Score'].value_counts())
```
