# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
# Write your code here
import pandas as pd
data = {
    'City': ['Bhopal', 'Indore', 'Bhopal', 'Delhi']
}

df = pd.DataFrame(data)

unique_cities = df['City'].unique()

print(unique_cities)
````


# Q2. Count Unique Values

Find the number of unique cities in a DataFrame.

## Sample Input

```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
# Write your code here

data = {
    'City': ['Bhopal', 'Bhopal', 'Delhi'],
    'Marks': [80, 90, 70]
}

df = pd.DataFrame(data)

avg_marks = df.groupby('City')['Marks'].mean()

print(avg_marks)
````


# Q5. Sort Values

## Problem Statement
Sort students by Marks in ascending order.

## Sample Output
```python
Sorted DataFrame
```


````python
# Write your code here

data = {
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [20, 22, 21],
    'City': ['New York', 'London', 'Paris'],
    'Marks': [85, 92, 78]
}

df = pd.DataFrame(data)

sorted_df = df.sort_values(by='Marks')

print(sorted_df)
````


# Q6. Sort Values Descending

## Problem Statement
Sort students by Marks in descending order.

## Sample Output
```python
Sorted DataFrame
```


````python
# Write your code here
sorted_df = df.sort_values(by='Marks', ascending=False)

print(sorted_df)
````


# Q7. Filter Records

## Problem Statement
Display students having marks greater than 75.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here

filtered_df = df[df['Marks'] > 75]

print(filtered_df)
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here

filtered_df = df[(df['Marks'] > 70) & (df['Age'] > 20)]

print(filtered_df)
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
# Write your code here

data = {
    'Employee': ['Amit', 'Rahul', 'Priya'],
    'Department': ['HR', 'IT', 'IT'],
    'Salary': [50000, 60000, 60000]
}

df = pd.DataFrame(data)

total_salary = df.groupby('Department')['Salary'].sum()

print(total_salary)
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
# Write your code here

data = {
    'Name': ['John', 'Alice', 'Bob', 'Emily'],
    'Age': [20, 22, 21, 23],
    'City': ['New York', 'London', 'Paris', 'London'],
    'Marks': [85, 92, 78, 95]
}

df = pd.DataFrame(data)

print(df['City'].unique())

print(df['City'].value_counts())

print(df[df['Marks'] > 80])

print(df.sort_values(by='Marks'))
````
